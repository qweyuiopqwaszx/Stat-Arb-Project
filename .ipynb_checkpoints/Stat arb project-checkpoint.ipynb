```python
import collections
from datetime import datetime
def executivesalary(tickers):
    salary = collections.defaultdict(list)
    for i in range(len(tickers)):
        info = yf.Ticker(tickers[i]).info
        companyOfficers = pd.DataFrame(info['companyOfficers'])
        try:
            salary[(tickers[i], 'totalpay')] = companyOfficers['totalPay'].mean()
        except:
            salary[(tickers[i], 'totalpay')] = None
        try:
            salary[(tickers[i], 'exercisedValue')] = companyOfficers['exercisedValue'].sum()
        except:
            salary[(tickers[i], 'exercisedValue')] = None
        salary[(tickers[i], 'trailingPegRatio')] = info['trailingPegRatio']
        try:
            salary[(tickers[i], 'targetMeanPrice')] = info['targetMeanPrice']
        except:
            salary[(tickers[i], 'targetMeanPrice')] = None
        try:
            salary[(tickers[i], 'next earning call')] = datetime.fromtimestamp(info['earningsTimestampStart'])
        except:
            salary[(tickers[i], 'next earning call')] = None
        salary[(tickers[i], 'shortRatio')] = info['shortRatio']
    return salary
    
```

```python
generalinfo = executivesalary(tickers)
```

```python
test = pd.Series(generalinfo)
```

```python
test2 = test.unstack(level = 1)
```
